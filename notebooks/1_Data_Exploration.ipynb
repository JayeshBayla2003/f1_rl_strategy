```python
#Plot LapTime against TyreLife. You should see a trend where lap times increase as the tires get older.
laps_ver = session.laps.pick_drivers(['VER'])
clean_laps_ver = laps_ver.pick_track_status('1', how='equals').pick_wo_box()

# Get the list of compounds he used in his clean laps
compounds = clean_laps_ver['Compound'].unique()
print(f"Verstappen's used compounds in clean laps: {compounds}")

# 4. DATA PREPARATION: Convert LapTime to seconds for plotting
# We do this for the entire DataFrame at once for efficiency.
clean_laps_ver['LapTimeSeconds'] = clean_laps_ver['LapTime'].dt.total_seconds()

# 5. PLOTTING
# Define a color for each compound to make the plot readable
compound_colors = {
    'SOFT': '#FF3333',   # Red
    'MEDIUM': '#FFF200', # Yellow
    'HARD': '#EBEBEB'    # White/Light Grey
}

plt.figure(figsize=(12, 8))
ax = plt.gca() # Get the current axes

for compound in compounds:
    # Filter the laps for the current compound
    compound_laps = clean_laps_ver[clean_laps_ver['Compound'] == compound]
    
    # Plot TyreLife vs LapTime for this compound
    ax.plot(
        compound_laps['TyreLife'], 
        compound_laps['LapTimeSeconds'], 
        color=compound_colors.get(compound, 'grey'),  # Use the defined color, or grey as a fallback
        linestyle='-', 
        marker='o',
        label=compound  # Label for the legend
    )

# 6. CUSTOMIZE THE PLOT
ax.set_title("Verstappen's Lap Times vs. Tyre Life\n2023 Bahrain Grand Prix")
ax.set_xlabel("Tyre Life (Laps)")
ax.set_ylabel("Lap Time (seconds)")
ax.invert_yaxis()  # Invert y-axis so faster laps are higher up
plt.suptitle("Trend shows lap times increasing as tires degrade", y=0.92)
ax.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Add a professional touch
plt.tight_layout()
plt.show()
```

```python
#Use a simple linear regression (from sklearn.linear_model.LinearRegression) to model this trend. The slope of this line is your degradation factor (e.g., 0.15 seconds per lap for a MEDIUM tire at Bahrain).
from sklearn.linear_model import LinearRegression

laps_alo = session.laps.pick_drivers(['ALO'])
clean_laps_alo = laps_alo.pick_track_status('1', how='equals').pick_wo_box()
hard_laps_alo = clean_laps_alo[clean_laps_alo['Compound'] == 'HARD']

# Convert lap time to seconds for numerical analysis
hard_laps_alo = hard_laps_alo.copy() # Avoid SettingWithCopyWarning
hard_laps_alo['LapTimeSeconds'] = hard_laps_alo['LapTime'].dt.total_seconds()

# 4. LINEAR REGRESSION
# The model requires X to be a 2D array and y to be a 1D array.
# X is our feature (TyreLife)
# y is our target (LapTimeSeconds)
X = hard_laps_alo['TyreLife'].values.reshape(-1, 1)
y = hard_laps_alo['LapTimeSeconds'].values

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Get the slope (degradation factor) and intercept
# The slope is the first (and only) coefficient in our simple model
degradation_factor = model.coef_[0]
intercept = model.intercept_

# Generate predicted lap times using the fitted model to draw the trend line
y_pred = model.predict(X)

# 5. PRINT THE RESULT
print(f"Driver: Fernando Alonso")
print(f"Tire Compound: HARD")
print(f"Calculated Degradation Factor (Slope): {degradation_factor:.4f} seconds/lap")
print(f"This means the lap time is expected to increase by ~{degradation_factor:.3f} seconds for each lap driven on the tire.")

# 6. PLOTTING
plt.figure(figsize=(12, 8))
ax = plt.gca()

# Plot actual data points
ax.scatter(X, y, color='white', edgecolor='black', zorder=3, label="Actual Lap Times")

# Plot the linear regression trend line
ax.plot(X, y_pred, color='red', linestyle='--', linewidth=2, label="Linear Regression Trend Line")

# Customize the plot
ax.set_title(f"Alonso's HARD Tyre Degradation\n2023 Bahrain GP (Degradation: {degradation_factor:.3f}s/lap)")
ax.set_xlabel("Tyre Life (Laps)")
ax.set_ylabel("Lap Time (seconds)")
ax.invert_yaxis() # Invert y-axis so faster laps are higher
ax.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Show the plot
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

import numpy as np
laps_per = session.laps.pick_drivers(['PER'])
clean_laps_per = laps_per.pick_track_status('1', how='equals').pick_wo_box()

# Add a 'LapTimeSeconds' column for numerical analysis
clean_laps_per['LapTimeSeconds'] = clean_laps_per['LapTime'].dt.total_seconds()

# 4. ANALYSIS LOOP
# Define the compounds to analyze and a dictionary to store results
compounds_to_analyze = ['SOFT', 'MEDIUM', 'HARD']
degradation_results = {}

for compound in compounds_to_analyze:
    # Filter for the specific compound
    compound_laps = clean_laps_per[clean_laps_per['Compound'] == compound]
    
    # A minimum number of laps is needed for a meaningful regression
    if len(compound_laps) < 3:
        print(f"Skipping {compound} compound: Not enough data points ({len(compound_laps)} laps).")
        continue

    # Prepare data for sklearn
    X = compound_laps['TyreLife'].values.reshape(-1, 1)
    y = compound_laps['LapTimeSeconds'].values
    
    # Create and fit the linear regression model
    model = LinearRegression()
    model.fit(X, y)
    
    degradation_factor = model.coef_[0]
    
    # Store results for plotting and summary
    degradation_results[compound] = {
        'model': model,
        'laps': compound_laps,
        'factor': degradation_factor
    }
    
    print(f"Degradation for {compound}: {degradation_factor:.4f} seconds/lap")

# 5. PLOTTING
plt.figure(figsize=(12, 8))
ax = plt.gca()

# Define colors for the plot
compound_colors = {'SOFT': '#FF3333', 'MEDIUM': '#FFF200', 'HARD': '#EBEBEB'}

for compound, results in degradation_results.items():
    laps = results['laps']
    model = results['model']
    factor = results['factor']
    
    # Scatter plot of actual lap times
    ax.scatter(
        laps['TyreLife'], 
        laps['LapTimeSeconds'], 
        color=compound_colors.get(compound), 
        edgecolor='black',
        label=f'{compound} Laps'
    )
    
    # Plot the regression trend line
    X_pred = np.arange(laps['TyreLife'].min(), laps['TyreLife'].max()).reshape(-1, 1)
    y_pred = model.predict(X_pred)
    ax.plot(
        X_pred, 
        y_pred, 
        color=compound_colors.get(compound), 
        linestyle='--', 
        linewidth=2,
        label=f'{compound} Trend (Deg: {factor:.3f}s/lap)'
    )

# Customize the plot
ax.set_title("Pérez's Tyre Degradation Analysis\n2023 Bahrain Grand Prix")
ax.set_xlabel("Tyre Life (Laps)")
ax.set_ylabel("Lap Time (seconds)")
ax.invert_yaxis()  # Faster laps are higher
ax.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Show the plot
plt.tight_layout()
plt.show()

# 6. FINAL SUMMARY
print("\n--- Summary of Degradation Factors ---")
for compound, results in degradation_results.items():
    print(f"{compound:<10}: {results['factor']:.4f} seconds per lap")

if 'MEDIUM' not in degradation_results:
    print("MEDIUM        : Not used by this driver in this session.")
```

```python
import os

# --- 1. SETUP ---
# Create a 'data' directory if it doesn't exist
output_dir = 'data'
os.makedirs(output_dir, exist_ok=True)
print("Calculating tyre degradation factors...")
laps_per = session.laps.pick_drivers(['PER'])
clean_laps_per = laps_per.pick_track_status('1', how='equals').pick_wo_box()
clean_laps_per['LapTimeSeconds'] = clean_laps_per['LapTime'].dt.total_seconds()

# Dictionary to store all our findings
analysis_results = {}

compounds_to_analyze = ['SOFT', 'MEDIUM', 'HARD']
for compound in compounds_to_analyze:
    compound_laps = clean_laps_per[clean_laps_per['Compound'] == compound]
    
    if len(compound_laps) < 3:
        continue

    X = compound_laps['TyreLife'].values.reshape(-1, 1)
    y = compound_laps['LapTimeSeconds'].values
    
    model = LinearRegression()
    model.fit(X, y)
    
    degradation_factor = model.coef_[0]
    
    # Store the result with a descriptive key
    analysis_results[f'{compound}_Degradation'] = degradation_factor

# --- 4. PIT STOP TIME LOSS ANALYSIS ---
print("Calculating average pit stop time loss...")

# Get all laps, not just clean ones
all_laps_per = session.laps.pick_drivers(['PER']).copy()
all_laps_per['LapTimeSeconds'] = all_laps_per['LapTime'].dt.total_seconds()

# Find the laps where the driver entered the pits
pit_in_laps = all_laps_per[all_laps_per['PitInTime'].notna()]
median_pit_lap_time = pit_in_laps['LapTimeSeconds'].median()

# Use the median of the clean HARD tire laps as a baseline for a normal racing lap
baseline_laps = clean_laps_per[clean_laps_per['Compound'] == 'HARD']
median_baseline_lap_time = baseline_laps['LapTimeSeconds'].median()

# Calculate the time loss
pit_stop_loss = median_pit_lap_time - median_baseline_lap_time
analysis_results['PitStopLoss'] = pit_stop_loss

# --- 5. SAVE FINDINGS TO CSV ---
# Convert the dictionary to a pandas DataFrame for easy saving
results_df = pd.DataFrame(list(analysis_results.items()), columns=['Metric', 'Value'])

# Define the output filename
output_filename = os.path.join(output_dir, '2023_Bahrain_PER_TyreAnalysis.csv')

# Save the DataFrame to a CSV file
results_df.to_csv(output_filename, index=False)
print(f"\nSuccessfully saved analysis results to: {output_filename}")

# --- 6. DISPLAY FINAL RESULTS ---
print("\n--- Summary of Findings ---")
for metric, value in analysis_results.items():
    if "Degradation" in metric:
        print(f"{metric:<20}: {value:.4f} seconds per lap")
    else:
        print(f"{metric:<20}: {value:.4f} seconds")
```
